```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('~/Downloads/query (62).csv', skiprows=3)
```

```python
# Get mean, std, p99, max for the column 'elapsed'
mean = df['elapsed'].mean()
std = df['elapsed'].std()
p99 = df['elapsed'].quantile(0.99)
max = df['elapsed'].max()

print(f'Mean: {mean}')
print(f'Std: {std}')
print(f'P99: {p99}')
print(f'Max: {max}')
```
